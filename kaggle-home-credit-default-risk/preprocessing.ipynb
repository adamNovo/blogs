```
def chart_distribution(df: pd.DataFrame):
    column_distributions = df.describe()
    idx = 0
    for i in column_distributions:
        fig = plt.figure()
        ax11 = fig.add_subplot(211)
        ax21 = fig.add_subplot(212)
        df_train_raw.loc[:,i].plot(kind="hist", bins=20, title=i, ax=ax11)
        column_distributions.loc[["min", "25%", "50%", "75%", "max", "mean"], i].plot(kind="bar", ax=ax21)
        print(idx, i)
        plt.show()
        idx += 1
chart_distribution(df_train_raw)
```

```
def chart_categorical(df: pd.DataFrame):
    idx = 0
    for col in df:
        if df[col].dtype == "object":
            df[col].value_counts().plot(kind="bar")
            print(idx, col)
            plt.show()
            idx += 1
chart_categorical(df_train_raw)
```

## Missing data

```
# train
def get_missing(df):
    exploration = pd.DataFrame(df.isnull().sum(), 
                 columns=["no_missing"])
    exploration = exploration.merge(
        pd.DataFrame(df.dtypes, columns=["type"]), 
        left_index=True, right_index=True)
    return exploration.loc[exploration["no_missing"]>0].sort_values(
        by="no_missing", ascending=False)
missing = get_missing(df_train)
for i, row in missing.iterrows():
    print(row)
```

```
# drop all columns with > n missing
print("Shape pre column drop: {}".format(df_train.shape))

df_shape = df_train.shape
for i, row in missing.iterrows():
    if row["no_missing"] > 140000 and i in df_train.columns:
        df_train = df_train.drop(i, axis=1)

print("Shape post column drop: {}".format(df_train.shape))
missing = get_missing(df_train)
for i, row in missing.iterrows():
    print(row)
```

```
# fill all objects using mode
fill_dict_object = {}
for i, row in missing.iterrows():
    if row["type"] == "object":
        fill_dict_object[i] = df_train[i].mode()[0]
df_train = df_train.fillna(fill_dict_object)

missing = get_missing(df_train)
for i, row in missing.iterrows():
    print(row)
```

```
# fill all remaining numerical using median
fill_dict_numerical = {}
for i, row in missing.iterrows():
    fill_dict_numerical[i] = df_train[i].median()
df_train = df_train.fillna(fill_dict_numerical)

missing = get_missing(df_train)
assert 0 == len(missing)
```

```
# test
missing = get_missing(df_test)
for i, row in missing.iterrows():
    print(row)
```

```
# drop columns dropped not in train
df_test = df_test.loc[:, list(df_train.columns)]

for i, row in get_missing(df_test).iterrows():
    print(row)
```

```
# fill missing using values from train
df_test = df_test.fillna(fill_dict_object)
df_test = df_test.fillna(fill_dict_numerical)
df_test = df_test.drop("TARGET", axis=1)

missing = get_missing(df_test)
assert 0 == len(missing)
```

## Correlated features

```
df_corr = df_train.corr()
show_threshold = 0.7
high_correlations = set()
for i, row in df_corr.iterrows():
    for j in row.index:
        if abs(df_corr.loc[i, j]) > show_threshold and i != j:
            print("{} | {} = {}".format(i, j, round(df_corr.loc[i, j], 4)))
            high_correlations.add(i)
            high_correlations.add(j)
```

```
high_correlations = list(high_correlations)
no_steps = 3
for i in range(len(high_correlations) // no_steps + 1):
    cols = high_correlations[i * no_steps:i * no_steps + no_steps]
    sns.pairplot(df_train.loc[:, cols], size=2)
    plt.show()
```
